# Vari√°veis aleat√≥rias discretas

S√£o vari√°veis do tipo quantitativas. 

Quando s√£o discretas significa que o intervalo entre dois n√∫meros n√£o √© considerado, ou seja, s√≥ existem n√∫meros inteiros como numa contagem de projetos, pe√ßas, filhos, vendas.

## Distribui√ß√£o binomial 

$Y$~$Bin(n,p)$

$$P(Y=y)=\left(
\begin{array}{c}
    n \\
    y
\end{array}\right) p^y (1-p)^{(n-y)}$$

Y = vari√°vel analizada (que pode ter sucesso ou fracasso)

y = quantas vezes acontece aquele sucesso

n = n√∫mero de tentativas, contagens, ensaios

p = probabilidade de sucesso em cada evento independente

## Condi√ß√µes para aplicar binomial:
- ser uma contagem de $n$ repeti√ß√µes (ou tentativas, ou ensaios) id√™nticas;
- cada repeti√ß√£o tem apenas 2 resultados poss√≠veis: um √© denominado sucesso (o resultado de interesse, que n√£o necessariamente √© positivo) e o outro, fracasso;
- a probabilidade de sucesso para cada ensaio √© denominada $p$ e ser√° constante em cada repeti√ß√£o. Consequentemente, a probabilidade de fracasso $(1-p)$ tamb√©m n√£o varia de tentativa para tentativa;
- as tentativas s√£o independentes.


**Esperan√ßa(o valor esperado, uma m√©dia ponderada):**
$$\overline{Y} = \sum\limits_y y\cdot P(Y=y)$$


**Vari√¢ncia:**
$$Var(Y) = \sum\limits_y (y-\overline{Y})^2\cdot P(Y=y)$$
    
**Desvio padr√£o:**
$$DP(Y) = \sqrt{Var(Y)}$$


```python
#Para exatamente 1 item
stats.binom.pmf(y,n,probabilidade_sucesso)
#Para o acumulo at√© o valor de y P(Y<=y)
stats.binom.cdf(y, n, probabilidade_sucesso)

#Encontrando a esperan√ßa(m√©dia ponderada)
#Forma 1
stats.binom.mean(n, p, loc=0)
#Forma 2
esperancateorica = 0 
for i in range(len(probabilidades_teoricas)):
    valor = i * probabilidades_teoricas[i]
    #i porque tenho 9 itens (n√∫mers de pe√ßas com sucesso), de 0 a 8
    esperancateorica += valor
    
print("A esperan√ßa √© de:",esperancateorica)
    
    
#Encontrando a vari√¢ncia
#Forma 1
stats.binom.var(n,probabilidade_sucesso)
#Forma 2
varianciateorica = 0
for i in range(len(probabilidades_teoricas)):
    valor = (i - esperancateorica)**2 * probabilidades_teoricas[i]
    varianciateorica += valor
    
print("A vari√¢ncia √© de:",varianciateorica)
```

# Teorema do Limite Central

A m√©dia das m√©dias das amostras (Xbarra) tem distribui√ß√£o normal
$$\bar{X}=\frac{X_1+X_2+...+X_n}{n}=\frac{\sum_{i=1}^n X_i}{n} \sim N\left(\mu,\frac{\sigma^2}{n}\right),$$



$$Z = \frac{\overline{X} - \mu}{\sigma/\sqrt{n}} \sim N\left(0,1\right).$$

<div id="passo1"></div>

Em geral a hip√≥tese alternativa denota algo not√°vel ou efeito an√¥malo que se queira testar (por exemplo resistores fora de conformidade)

### Passo 2

Defina qual estat√≠stica ser√° usada para julgar a hip√≥tese $H_0$.

Lembre-se:

* Estat√≠stica √© calculada na **amostra**
* A hip√≥tese diz respeito √† **popula√ß√£o** de onde veio a amostra

**Importante:** usar $\overline{x}$ (valor estimado na amostra) e  a distribui√ß√£o de $\overline{X}$ (distribui√ß√£o de m√©dias amostrais) para inferir (tomar decis√£o) algo sobre $\mu$ (m√©dia populacional).

<div id="passo3"></div>

### Passo 3

Decida qual $\alpha$ utilizar e defina a Regi√£o Cr√≠tica (RC) segundo $H_0$.

Sobre $\alpha$:

* $\alpha$ √© a probabilidade de rejeitar $H_0$ mesmo que ela seja verdadeira.
* $\alpha$ √© chamado de n√≠vel de signific√¢ncia.
* Tipicamente $\alpha$ √© fixado em $1\%$, $5\%$ ou $10\%$ 
* $1 - \alpha$  √© chamado de n√≠vel de confian√ßa (ou coeficiente de confian√ßa).

#### Exemplos de regi√£o cr√≠tica:

a. Direita:

$H_0: \mu = \mu_0$

$H_1: \mu > \mu_0$

$x_{cr√≠tico}$ √† direita

b. Esquerda 

Para:

$H_0: \mu = \mu_0$

$H_1: \mu < \mu_0$

$x_{cr√≠tico}$ √† esquerda

c. Bicaudal 

Para: 

$H_0: \mu = \mu_0$

$H_1: \mu \neq \mu_0$

$x_{cr√≠tico}$ em ambas as caudas

# Regress√£o Linear

## Como analisar a regress√£o linear 
* *$R^2$:* Este valor √© uma medida de ajustamento de um modelo estat√≠stico linear generalizado varia entre 0 e 1, indicando, em percentagem, o quanto o modelo consegue explicar os valores observados. Quanto maior o R¬≤, mais explicativo √© o modelo, melhor ele se ajusta √† amostra.
* **Coef:** s√£o os coeficientes da regress√£o linear, o **coef constantes** √© const
* **Pvalue:** Este valor varia de 0 a 1, quanto mais proximo do 0 mais correlacionado s√£o os dados
* **F-statistic:** Seria uma jun√ß√£o dos Pvalue, analisa todos os Pvalue, ou seja, ele da a probabilidade dos coeficientes serem encontrados como √ü1...√ün mas na realidade n√£o era para existir
    * **Este varia de 0 a 1**: quanto mais perto do um, menos correlacionados s√£o os dados, estes betas n√£o deveriam existir
    * **Testa as seguintes hip√≥teses**:
        * H0:Œ≤1=Œ≤2=‚Ä¶=Œ≤p=0
        * H1:Œ≤i‚â†0 para no m√≠nimo algum  i, com  i=1,‚Ä¶,n
 
    * A rejei√ß√£o da hip√≥tese nula implica que no m√≠nimo uma vari√°vel explicativa (regressora) contruibui significantemente para o modelo. Esse teste F √© mais √∫til no caso de modelo de regress√£o m√∫ltipla.
* **Durbin-Watson:** √© utilizado para detectar a presen√ßa de autocorrela√ß√£o (depend√™ncia) nos res√≠duos de uma an√°lise de regress√£o. Ele varia de 0 a 4, sendo que de 0 a 2 indica que h√° uma autocorrela√ß√£o positiva e de 2 a 4 indica uma autocorrela√ß√£o negativa. 
    * Um valor pr√≥ximo de  2 sugere que n√£o h√° autocorrela√ß√£o dos res√≠duos
    * Um valor menor que  2 sugere correla√ß√£o positiva dois res√≠duos
    * Um valor maior que  2 sugere correla√ß√£o negativa dois res√≠duos
* **Omnibus:** testa a normalidade dos residuos
* **Prob(Omnibus):** for muito baixo, existe evid√™ncia de que os res√≠duos n√£o s√£o distribu√≠dos normalmente, violando nesse caso a suposi√ß√£o do modelo de regress√£o.
    * Caso o enunciado n√£o dizer nada, se a probabilidade de ombibus for maior que 5% j√° podemos concluir que segue uma distribui√ß√£o normal
* **JB**: testa a normalidade dos residuos
* **Prob(JB):** for muito baixo, existe evid√™ncia de que os res√≠duos n√£o s√£o distribu√≠dos normalmente, violando nesse caso a suposi√ß√£o do modelo de regress√£o.
